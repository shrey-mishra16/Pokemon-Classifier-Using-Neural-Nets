```python
class NeuralNetwork:
    def __init__(self,inp,layer,op):
        np.random.seed(0)
        model={}
        
        model['W1'] = np.random.randn(inp,layer[0])
        model['B1'] = np.zeros((1,layer[0]))
                
        model['W2'] = np.random.randn(layer[0],layer[1])
        model['B2'] = np.zeros((1,layer[1]))
        
        model['W3'] = np.random.randn(layer[1],layer[2])
        model['B3'] = np.zeros((1,layer[2]))
        
        model['W4'] = np.random.randn(layer[2],layer[3])
        model['B4'] = np.zeros((1,layer[3]))
                
        model['W5'] = np.random.randn(layer[3],op)
        model['B5'] = np.zeros((1,op))
        
        
        self.model= model
        self.activation_outputs=None

    def fwd(self,x):
        
        W1,W2,W3,W4,W5= self.model['W1'],self.model['W2'],self.model['W3'],self.model['W4'],self.model['W5']
        b1,b2,b3,b4,b5= self.model['B1'],self.model['B2'],self.model['B3'],self.model['B4'],self.model['B5']
        
        z1=np.dot(x,W1)+b1
        a1=np.tanh(z1)
        
        z2=np.dot(a1,W2)+b2
        a2=np.tanh(z2)
        
        z3=np.dot(a2,W3)+b3
        a3=np.tanh(z3)
        
        z4=np.dot(a3,W4)+b4
        a4=np.tanh(z4)
        
        z5=np.dot(a4,W5)+b5
        y_=softmax(z5)
        
        self.activation_outputs=(a1,a2,a3,a4,y_)
        return y_
    
    def backward(self,x,y,learning_rate=0.0003):
        W1,W2,W3,W4,W5= self.model['W1'],self.model['W2'],self.model['W3'],self.model['W4'],self.model['W5']
        b1,b2,b3,b4,b5= self.model['B1'],self.model['B2'],self.model['B3'],self.model['B4'],self.model['B5']
        m = x.shape[0]
        
        a1,a2,a3,a4,y_ = self.activation_outputs
        
        delta5 = y_ - y
        dw5 = np.dot(a4.T,delta5)
        db5 = np.sum(delta5,axis=0)
        
        delta4 = (1-np.square(a4))*np.dot(delta5,W5.T)
        dw4 = np.dot(a3.T,delta4)
        db4 = np.sum(delta4,axis=0)
        
        delta3 = (1-np.square(a3))*np.dot(delta4,W4.T)
        dw3 = np.dot(a2.T,delta3)
        db3 = np.sum(delta3,axis=0)
        
        delta2 = (1-np.square(a2))*np.dot(delta3,W3.T)
        dw2 = np.dot(a1.T,delta2)
        db2 = np.sum(delta2,axis=0)
        
        delta1 = (1-np.square(a1))*np.dot(delta2,W2.T)
        dw1 = np.dot(x.T,delta1)
        db1 = np.sum(delta1,axis=0)
        
        
        #Update the Model Parameters using Gradient Descent
        self.model["W1"]  -= learning_rate*dw1
        self.model['B1']  -= learning_rate*db1
        
        self.model["W2"]  -= learning_rate*dw2
        self.model['B2']  -= learning_rate*db2
        
        self.model["W3"]  -= learning_rate*dw3
        self.model['B3']  -= learning_rate*db3

        self.model["W4"]  -= learning_rate*dw4
        self.model['B4']  -= learning_rate*db4
        
        self.model["W5"]  -= learning_rate*dw5
        self.model['B5']  -= learning_rate*db5
        
    def predict(self,x):
        y_out = self.fwd(x)
        return np.argmax(y_out,axis=1)
    
    def summary(self):
        W1,W2,W3,W4,W5= self.model['W1'],self.model['W2'],self.model['W3'],self.model['W4'],self.model['W5']
                
        a1,a2,a3,a4,y_ = self.activation_outputs
        
        print("W1 ",W1.shape)
        print("A1 ",a1.shape)       
        
def softmax(a):
    e_pa = np.exp(a) #Vector
    ans = e_pa/np.sum(e_pa,axis=1,keepdims=True)
    return ans        
```

```python
def loss(y_oht,p):
    l = -np.mean(y_oht*np.log(p))
    return l

def one_hot(y,depth):
    
    m = y.shape[0]
    y_oht = np.zeros((m,depth))
    y_oht[np.arange(m),y] = 1
    return y_oht
```

```python
def train(X,Y,model,epochs,learning_rate,logs=True):
    training_loss = []
    
    classes = len(np.unique(Y))
    Y_OHT = one_hot(Y,classes)
    
    for ix in range(epochs):
        
        Y_ = model.fwd(X)
        l = loss(Y_OHT,Y_)
        
        model.backward(X,Y_OHT,learning_rate)
        training_loss.append(l)
        if(logs and ix%50==0):
            print("Epoch %d Loss %.4f"%(ix,l))
            
    
    return training_loss
```
